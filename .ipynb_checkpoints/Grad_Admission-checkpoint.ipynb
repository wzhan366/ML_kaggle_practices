1. GRE Scores ( out of 340 ) 
2. TOEFL Scores ( out of 120 ) 
3. University Rating ( out of 5 ) 
4. Statement of Purpose and Letter of Recommendation Strength ( out of 5 ) 
5. Undergraduate GPA ( out of 10 ) 
6. Research Experience ( either 0 or 1 ) 
7. Chance of Admit ( ranging from 0 to 1 )

# pre-process

```python
def df_transform(_df): 
    '''
    Use this funtion to drop cols
    '''
    train_df_without_num = _df.drop(labels=['Num'],axis=1)
    t_df = train_df_without_num.drop(labels=['Research'],axis=1)
    return t_df
```

```python
train_df = df_transform(train_df)
train_df.head()
```

```python
train_coor = train_df.corr()
print train_coor
```

```python
test_df = df_transform(test_df)
test_df.head()
```

```python
def split_label(_df):
    X_ = _df.iloc[:,:-1]
    y_ = _df.iloc[:,-1]
    return X_, y_
```

```python
train_X, train_y = split_label(train_df)
test_X, test_y = split_label(test_df)
```

#### Key parameters
